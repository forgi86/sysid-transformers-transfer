```python
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge

# Create a figure and axis
figsize = set_size(linewidth, fraction=1.0, ratio=1.0)

# Create a figure and axis
fig, ax = plt.subplots(figsize=figsize)

# Define the center and radius of the disk
center = (0, 0)  # (x, y)
radius = 1.0

# Plot the disk using a circle patch
#circle = plt.Circle(center, radius, color='black', alpha=0.05)  # You can change the color and transparency
#ax.add_patch(circle)
#theta = np.linspace(0, 2*np.pi, 1000)
#plt.plot(np.cos(theta), np.sin(theta), "k", label="unit circle")


# Define the sector of the disk to color
start_angle_a = -90  # Start angle in degrees
end_angle_a = 90    # End angle in degrees
inner_radius_a = 0.8 * radius
outer_radius_a = 0.97 * radius

start_angle_b = 90  # Start angle in degrees
end_angle_b = 270    # End angle in degrees
inner_radius_b = 0.5 * radius
outer_radius_b = 0.75 * radius

start_angle_c = 0  # Start angle in degrees
end_angle_c = 360    # End angle in degrees
inner_radius_c = 0.5 * radius
outer_radius_c = 0.97 * radius


# Create a wedge-shaped patch for the sector
sector_a = Wedge(center, outer_radius_a, start_angle_a, end_angle_a, width=outer_radius_a - inner_radius_a, color='C0', label="a")#, alpha=0.99)
sector_b = Wedge(center, outer_radius_b, start_angle_b, end_angle_b, width=outer_radius_b - inner_radius_b, color='C1', label="b")#, alpha=0.99)
sector_c = Wedge(center, outer_radius_c, start_angle_c, end_angle_c, width=outer_radius_c - inner_radius_c, label="c", alpha=0.2, facecolor="gray", hatch="++", edgecolor="k")


# Add the sector to the plot
ax.add_patch(sector_a)
ax.add_patch(sector_b)
ax.add_patch(sector_c)


# Set the aspect ratio to be equal to ensure the disk and sector appear correctly
ax.set_aspect('equal')

# Set the limits of the plot
ax.set_xlim(-1.6, 1.6)
ax.set_ylim(-1.6, 1.6)

#ax.set_xlim(-1.2, 1.2)
#ax.set_ylim(-1.2, 1.2)

# Add labels and a title (optional)
ax.set_xlabel('Real part')
ax.set_ylabel('Imaginary part')
#ax.set_title('Pole distribution')
#plt.legend(bbox_to_anchor=(1.5, 1.03))
plt.legend(loc="upper right")

plt.grid(True)
#plt.grid(False)
plt.tight_layout()
plt.savefig("pole_distribution.pdf");
#plt.savefig("pole_distribution.png");
```
